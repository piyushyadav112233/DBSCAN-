```python
#Use ACF and PACF plots to determine the parameters.
# Plot ACF and PACF
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(df['Usage'], lags=40, ax=ax1)

# Calculate the maximum allowed lags (50% of sample size)
max_lags = len(df['Usage']) // 2

# Use the minimum between the desired lags (40) and the maximum allowed lags
plot_pacf(df['Usage'], lags=min(40, max_lags), ax=ax2)

plt.show()
```

```python
#2. Model Fitting
#Fit the ARIMA model using the selected parameters (p, d, q).
model=ARIMA(df['Usage'],order=(1,1,1))
model_fit=model.fit()
model_fit.summary()
```

```python
#plot
model_fit.plot_diagnostics(figsize=(12, 8))
plt.show()
```

```python
#Forecasting
#Perform out-of-sample forecasting.
forecast_steps=10
forecast=model_fit.forecast(steps=forecast_steps)
forecast
```

```python
#plot
plt.figure(figsize=(12, 6))
plt.plot(df['Usage'], label='Actual Usage')
plt.plot(pd.date_range(start=df['Date'].max(), periods=forecast_steps, freq='D'), forecast, label='Forecasted Usage', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Usage')
plt.title('Actual vs. Forecasted Usage')
plt.legend()
plt.show()
```

```python
model=ExponentialSmoothing(df['Usage'],trend='add',seasonal='add',seasonal_periods=7)
model_fit=model.fit()
model_fit.summary()
```

```python
#2. Parameter Optimization
#Use AIC for parameter optimization.
print(model_fit.aic)
print(model_fit.params)
print(model_fit.fittedvalues)
```

```python
print(f'AIC: {model_fit.aic}')
print(f'BIC: {model_fit.bic}')
print(f'SSE: {model_fit.sse}')
```

```python
 #Model Fitting and Forecasting
#Fit the model and visualize the forecasts.
forecast_steps = 10
forecast = model_fit.forecast(steps=forecast_steps)
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Usage'], label='Actual Usage')
plt.plot(pd.date_range(start=df['Date'].max(), periods=forecast_steps, freq='D'), forecast, label='Forecasted Usage', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Usage')
plt.title('Actual vs. Forecasted Usage')
plt.legend()
plt.show()
```

```python
#valuation and Comparison
#Calculate MAE, RMSE, and MAPE.
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(df['Usage'], model_fit.fittedvalues)

# Calculate RMSE without the 'squared' argument and then take the square root
mse = mean_squared_error(df['Usage'], model_fit.fittedvalues)  # Calculate MSE
rmse = np.sqrt(mse)  # Calculate RMSE by taking the square root of MSE

mape = np.mean(np.abs((df['Usage'] - model_fit.fittedvalues) / df['Usage'])) * 100
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
print(f'MAPE: {mape}')
```
